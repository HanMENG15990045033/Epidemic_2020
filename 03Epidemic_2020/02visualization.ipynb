## 热力图

```python
def Visualization(UpdateData):
#     latitude = np.array(UpdateData["latitude"])
    latitude = UpdateData["latitude"].values
    longitude = np.array(UpdateData["longitude"])
    confirmed = np.array(UpdateData["confirmed"],dtype=float)
    
    Vis_Data = [[latitude[i],longitude[i],confirmed[i]] for i in range(len(UpdateData))]
    
    map_osm = folium.Map(location=[30,114],zoom_start=4) # 新建地图，中心，缩放程度
    HeatMap(Vis_Data).add_to(map_osm) 
    return map_osm   
```

```python
map_osm = Visualization(Data)
map_osm
```

```python
file_path = r"Visualization.html"
map_osm.save(file_path)   #保存本地
webbrowser.open(file_path) #在本地浏览器打开
```

```python
cityName = Data["cityName"].values
latitude = Data["latitude"].values
longitude = Data["longitude"].values
cured = Data["cured"].values
confirmed = Data["confirmed"].values
confirmed = confirmed + cured
dead = Data["dead"].values
potential = dead / 0.015
r_b = np.power(potential, 0.5)*3000
r_g = np.power(confirmed, 0.5)*3000
r_r = np.power(dead, 0.5)*3000
```

```python
color_piker = {'r':'#cc0000', 'y':'#ff9900', 'b':'#3399ff'}
```

```python
import folium

# 中国地图图层
m = folium.Map(location=[30.52, 114.31], zoom_start=4)

# 潜在数
for i in range(len(Data)):
    folium.Circle(
    radius=r_b[i],
    location=[latitude[i], longitude[i]],
    tooltip= "{0}潜在{1}例".format(cityName[i], int(potential[i])),    
    color=color_piker['b'],
    fill=True,
    fill_color = color_piker['b']
).add_to(m)
    
# 确诊数
for i in range(len(Data)):
    folium.Circle(
    radius=r_g[i],
    location=[latitude[i], longitude[i]],
    tooltip= "{0}确诊{1}例".format(cityName[i], int(confirmed[i])),    
    color=color_piker['y'],
    fill=True,
    fill_color = color_piker['y']
).add_to(m)


# 死亡
for i in range(len(Data)):
    folium.Circle(
    radius=r_r[i],
    location=[latitude[i], longitude[i]],
    tooltip="{0}死亡{1}例".format(cityName[i], int(dead[i])),
    color='#ff6666',
    fill=True,
    fill_color = color_piker['r']
    ).add_to(m)

m
```

```python
# 保存为html
m.save('quanquantu.html')
```

先做一个自动聚类吧

这个图适合每个人都有一个坐标

```python
Data.sum()
```

城市坐标+随机数

模拟每个人的坐标

纬线是111KM/度，经线在赤道是111KM，其余的是111XCOS(A),A为所在纬度数

```python
 Data['latitude'][0]
```

```python
import random
random.uniform(0, 0.1)
```

```python
lats = np.array([])
lons = np.array([])
```

```python
Data['latitude'].values[0]
```

```python
Data['longitude'].values[0]
```

```python
Data['confirmed'].values[0]
```

```python
for i in range(len(Data)):
    center = Data['latitude'].values[i] # 以城市坐标为中心
    num = Data['confirmed'].values[i] # 确诊人数
    for j in range(num):
        lats = np.append(lats, center + random.uniform(0,0.2))
    print(i, j)
```

```python
for i in range(len(Data)):
    center = Data['longitude'].values[i]
    num = Data['confirmed'].values[i] # 确诊人数
    for j in range(num):
        lons = np.append(lons, center + random.uniform(0,0.2))
```

```python
len(lons)
```

```python
from folium.plugins import FastMarkerCluster


m = folium.Map(
    location=[np.mean(lats), np.mean(lons)],
    zoom_start=3
)

FastMarkerCluster(data=list(zip(lats, lons))).add_to(m)

folium.LayerControl().add_to(m)

m
```

```python
file_path = r"Visualization_cluster.html"
m.save(file_path)   #保存本地
webbrowser.open(file_path) #在本地浏览器打开
```
